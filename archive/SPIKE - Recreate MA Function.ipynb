```python
from datetime import datetime

import pandas as pd


def read_data(ticker = "BTC", timeframe="1W") -> pd.DataFrame:
    _data = pd.read_csv(f"../../data/INDEX_{ticker}USD_{timeframe}.csv")
    _data.rename(
        columns={
            "MA": "MA100",
            #"ATR": "ATR1",
            #"ATR.1": "ATR14",
            "MA.1": "MA8",
            "MA.2": "MA50",
            "20w SMA": "MA20",
            "21w EMA": "EMA21",
            #"RSI-based MA": "RSI MA",
            "time": "DateTime",
        }, 
        inplace=True
    )
    _data = _data[
        [
            "DateTime",
            "open",
            "high",
            "low",
            "close",
            "Volume",
            #"Volume MA",
            "MA100",
            "MA50",
            "EMA21",
            "MA20",
            "MA8",
            #"RSI",
            #"RSI MA",
            #"ATR1",
            #"ATR14",
        ]
    ]
    _data["DateTime"] = _data["DateTime"].apply(
        lambda row: datetime.fromtimestamp(row)
    )

    return _data
```

```python
import numpy as np


def calculate_moving_average(close_price, ma_length):
    weights = np.ones(ma_length) / ma_length
    print(weights)
    moving_average = np.convolve(close_price, weights, mode="valid")
    moving_average_with_nans = np.concatenate((np.full(ma_length - 1, np.nan), moving_average))
    return moving_average_with_nans
```

```python
from datetime import datetime, timedelta
from typing import Tuple

def _start_and_end_date(interval: str) -> Tuple[datetime, datetime]:
    ALLOWABLE_TIMES = {
        "long_term": ["1d", "1wk", "1mo"],
        "mid_term": ["1h"],
        "short_term": ["5m", "15m", "30m"],
        "shorter_term": ["1m"],
    }
    end = datetime.now()
    if interval in ALLOWABLE_TIMES["long_term"]:
        start = None
    elif interval in ALLOWABLE_TIMES["mid_term"]:
        start = end - timedelta(730)
    elif interval in ALLOWABLE_TIMES["short_term"]:
        start = end - timedelta(60)
    elif interval in ALLOWABLE_TIMES["shorter_term"]:
        start = end - timedelta(7)
    else:
        raise ValueError(f"Time step provided ({interval}) not valid {list(ALLOWABLE_TIMES.values())}")
            
    return start, end

def get_yahoo_data(ticker: str = "BTC-USD", interval: str = "1wk") -> pd.DataFrame:
    start, end = _start_and_end_date(interval)

    data = yf.download(
        tickers = ticker,
        interval = interval,
        start = start,
        end = end,
    )

    return data
```
